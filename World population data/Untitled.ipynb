```python
#Importing the requirements
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
df = pd.read_csv("world_population_data.csv")
df
```

```python
#cleanining the data
df.isna().sum() #found no null values
```

```python
#checking the duplication
df.duplicated().sum()
```

```python
sns.histplot(df['2023 population'], bins=10)
plt.title("Distribution of the 2023 populations")
plt.xlabel("Population")
plt.ylabel("frequency")
plt.show()
```

```python
#groupby the continent
df.groupby('continent')['2023 population'].mean().reset_index()
# df.groupby('continent')['2023 population'].mean().max()
```

```python
#bar plot
top_10_countries= df.sort_values(by = '2023 population', ascending= False).head(10)
```

```python
top_10_countries
```

```python
#bar plot
sns.barplot(
    x='2023 population',
    y='country',
    data=top_10_countries,
    hue='country',
    legend=False,
    palette='viridis'
)
plt.title("bar plot for the country vs population of 2023")
plt.xlabel("population")
plt.ylabel("country")
plt.show()
```

```python
#scatter plot
sns.scatterplot(x = '2023 population',y='country',data= top_10_countries,hue='country')
plt.show()
```

```python
#rename the name of the column
df.rename(columns = {'density (km²)': 'density(km2)', 'area (km²)': 'area(km2)'},inplace = True)
df.columns
```

```python
#scatter plot for population vs the area(km2)
sns.scatterplot(x = '2023 population',y= 'area(km2)', data= top_10_countries, hue='country')
plt.title("Population and the area scatter plot")
plt.xlabel("Population")
plt.ylabel("area(km2)")
plt.show()
```
