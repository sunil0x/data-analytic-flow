```python
#find the number of times the weather is exactly clear.
count_clear = df[df['Weather'] == 'Clear'].shape[0]
print(count_clear)
```

```python
#using the sum
count_clear = (df['Weather'] == 'Clear').sum()
print(count_clear)
```

```python
#3. Find the number of times when the wind speed was exactly 4 kmh
count = df[df['Wind Speed'] == 4].value_counts().shape[0]
count
```

```python
#mean value of the each columns against each 'Weather condition'
df.groupby('Weather_condition').mean(numeric_only = True) #considering the numeric column only
```

```python
#minimum and maximum value of the each column against each weather condition
print(df.groupby('Weather_condition').max())
print(df.groupby('Weather_condition').max())
```
