```python
#reading the new_dataset
df_new = pd.read_excel('Country-code.xlsx')
df_new
```

```python
df_new.columns
```

```python
df.shape
```

```python
#merging the two dataset
df_merged= pd.merge(df, df_new, on='Country Code', how='left')
df_merged
```

```python
df_final = df_merged.copy()
```

```python
df_final.head(2)
```

```python
df_final.columns
```

```python
country_name = df_final.Country.value_counts().index #max number of transactions is happenning in india
```

```python
country_val = df_final.Country.value_counts().values
```

```python
#pie chart - top3
plt.pie(country_val[:3], labels = country_name[:3], autopct="%1.2f%%")
plt.title("top 3 most transactions country for the zomato")
plt.show()
```

### Observation from the piechart: Zomato maximum or transactions are from india, usa and united kingdom

```python
df_final.columns
```

```python
ratings =df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
ratings
```

```python
#observations:
# 1. when rating is betn the 4.5 to 4.9-> Excellent
# 2. 1.8- 2.4 -> Poor
# 3. 2.6- 3.4 -> average
# 4. 3.5 - 4 -> Good
# 5. 4 to 4.4 -> Very good

#many people havenot rate anything i.e. 2148
```

```python
ratings.head()
```

```python
#rename the 0 as the rating count
ratings.rename(columns = {0:'rating count'}, inplace= True)
ratings.columns
```

```python
#bar plot for aggregate rating vs the rating count
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6) #adjusting the size of the image
sns.barplot(x = 'Aggregate rating', y= 'rating count', data= ratings)
plt.title("Aggregate rating vs the rating count")
plt.show()
```

```python
sns.barplot(x = 'Aggregate rating', y= 'rating count', hue='Rating color',data= ratings
           ,palette = ['blue','red','orange','yellow','green','green'])
plt.show()
```

```python
#Observations: 1. not rated count is very high
#2. max ratings are betn the 2.5 and 4

```

```python
#count plot
sns.countplot(x= 'Rating color', data = ratings, palette= ['blue','red','orange','yellow','green','green'])
plt.show()
```
